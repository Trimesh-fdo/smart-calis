Explore the synthetic dataset generated for training the three ML models.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid', palette='muted')
plt.rcParams['figure.figsize'] = (12, 5)

df = pd.read_csv('../data/raw/calisthenics_dataset.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
df.info()
```

```python
df.describe().T.style.background_gradient(cmap='Blues')
```

```python
# Missing values
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0] if missing.any() else 'None — dataset is complete.')
```

## Categorical Distributions

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 4))
cats = ['gender', 'goal', 'fitness_level', 'intensity']
for ax, col in zip(axes, cats):
    counts = df[col].value_counts()
    ax.bar(counts.index, counts.values, color=sns.color_palette('muted', len(counts)))
    ax.set_title(col.replace('_', ' ').title())
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## Numeric Distributions

```python
nums = ['age', 'height_cm', 'weight_kg', 'duration_min', 'sets', 'reps', 'calories_burned']
df[nums].hist(bins=30, figsize=(18, 8), color='steelblue', edgecolor='white')
plt.suptitle('Numeric Feature Distributions', y=1.01, fontsize=14)
plt.tight_layout()
plt.show()
```

## Exercise Frequency by Goal

```python
pivot = df.groupby(['goal', 'exercise']).size().unstack(fill_value=0)
pivot.plot(kind='bar', stacked=True, figsize=(12, 5), colormap='tab10')
plt.title('Exercise Frequency by Goal')
plt.ylabel('Count')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Calories Burned — Goal vs Intensity

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df.boxplot(column='calories_burned', by='goal', ax=axes[0])
axes[0].set_title('Calories by Goal')
axes[0].set_xlabel('')
df.boxplot(column='calories_burned', by='intensity', ax=axes[1])
axes[1].set_title('Calories by Intensity')
axes[1].set_xlabel('')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

## Correlation Heatmap

```python
corr = df[nums].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(9, 7))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            linewidths=0.5, cbar_kws={'shrink': 0.8})
plt.title('Numeric Feature Correlation')
plt.tight_layout()
plt.show()
```
